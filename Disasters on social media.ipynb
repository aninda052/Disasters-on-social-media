# Disasters on social media


Contributors looked at over 10,000 tweets culled with a variety of searches like “ablaze”, “quarantine”, and “pandemonium”, then noted whether the tweet referred to a disaster event (as opposed to a joke with the word or a movie review or something non-disastrous).

***This data set was collected from Crowdflower.***

We will try to correctly predict tweets that are about disasters

```python
import codecs
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Reading & Cleaning

```python
data=codecs.open('socialmedia-disaster-tweets-DFE.csv',encoding='utf8',errors='replace')
file=open('nonUnicode_clean.csv','w')
for line in data:
    file.write(line)
file.close()
    
```

```python
data=pd.read_csv('nonUnicode_clean.csv')
print(data.head())
```

```python
data['class_label']=[1 if a=='Relevant' else 0 if a=='Not Relevant'else 2 for a in data['choose_one']]
```

```python
data=data[['text','choose_one','class_label']]
```

```python
def cleaning_text(df, col_name):
    df[col_name] = df[col_name].str.replace(r"http\S+", "").replace(r"http", "")\
        .replace(r"@\S+", "").replace(r"[^A-Za-z0-9(),!?@\'\`\"\_\n]", " ")\
        .replace(r"@", "at")
    df[col_name] = df[col_name].str.replace(r"#", "")
    df[col_name] = df[col_name].str.lower()
    return df

data = cleaning_text(data, "text")
```

```python
print(data.head())
```

```python
data.to_csv('proper_clean_data.csv') #for further use
```

# Tokenizing

```python
from nltk.tokenize import RegexpTokenizer
```

```python
token = RegexpTokenizer(r'\w+')
data["tokens"] = data["text"].apply(token.tokenize)
data.head()
```

# Enter embeddings

Our goal is to first create a useful embedding for each sentence (or tweet) in our dataset, and then use these embeddings to accurately predict the relevant category.


```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
text=data["text"].tolist()
label=data["class_label"].tolist()
x_train,x_test,y_train,y_test=train_test_split(text,label,test_size=.30,random_state=5)
```


# Bag of Words Counts

```python
cnt_vec=CountVectorizer()
x_train_cnt=cnt_vec.fit_transform(x_train)
x_test_cnt=cnt_vec.transform(x_test)
```

```python
x_train_cnt.shape
```

```python
from sklearn.decomposition import TruncatedSVD
import matplotlib.colors as col
import matplotlib.patches as mpatches
```

# Visualizing the embeddings

Now that we've created embeddings, let's visualize them and see if we can identify some structure.Since visualizing data in allmost 17k dimensions is hard, let's project it down to 2.

```python
def plot_LSA(data,labels):
        lsa = TruncatedSVD(n_components=2)
        lsa_scores = lsa.fit_transform(data)
        colors = ['DarkOrange','DarkGreen','DarkOrange']
        plt.figure(figsize=(10, 10)) 
        plt.scatter(lsa_scores[:,0], lsa_scores[:,1], s=8, alpha=.8, c=labels, cmap=col.ListedColormap(colors))
        red_patch = mpatches.Patch(color='DarkOrange', label='Irrelevant')
        green_patch = mpatches.Patch(color='DarkGreen', label='Disaster')
        plt.legend(handles=[red_patch, green_patch], prop={'size': 30})
        plt.show()
```

```python
plot_LSA(x_train_cnt,y_train)
```

These embeddings don't look very cleanly separated. Let's see if we can still fit a useful model on them

# Fitting a classifier

We will use logistic regression . It is simple, often gets the job done, and is easy to interpret.

```python

from sklearn.linear_model import LogisticRegression
```

```python
model_cnt=LogisticRegression(C=30.0, class_weight='balanced', solver='newton-cg', 
                         multi_class='multinomial', n_jobs=-1, random_state=5)
model_cnt.fit(x_train_cnt, y_train)
predicted_cnt=model_cnt.predict(x_test_cnt)
```

# Evaluation


from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import numpy as np
import itertools

```python
print(classification_report(y_test,predicted_cnt))
print('Accuracy: %.2f%%'%(accuracy_score(y_test,predicted_cnt)*100))
```

```python
def plot_confusion_matrix(cm, classes,title,cmap=plt.cm.Set2):
    a=plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=30)
    plt.colorbar(fraction=.045)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, fontsize=20)
    plt.yticks(tick_marks, classes, fontsize=20)
    
    fmt ='d'
    thresh = cm.max() / 2.

    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", 
                 color="white" if cm[i, j] < thresh else "black", fontsize=40)
    
    plt.tight_layout()
    plt.ylabel('True label', fontsize=30)
    plt.xlabel('Predicted label', fontsize=30)

    return plt

cm = confusion_matrix(y_test,predicted_cnt)
fig = plt.figure(figsize=(10, 10))
plot = plot_confusion_matrix(cm, classes=['Irrelevant','Disaster','Unsure'],title='Confusion matrix')
plt.show()
```

Our classifier never predicts class 3, which is not surprising, seeing as it is critically undersampled. This is not very important here, as the label is not very meaningful. Our classifier creates more false negatives than false positives . Depending on the use case, this seems desirable.

# Further inspection

Let's look at the features our classifier is using to make decisions

```python

def important_features(vectorizer, model, n=5):
    index_to_word = {idx:word for word,idx in vectorizer.vocabulary_.items()}

    classes ={}
    for class_idx in range(model.coef_.shape[0]):
        word_importances = [(val, index_to_word[word_idx]) for word_idx,val in enumerate(model.coef_[class_idx])]
        sorted_coeff = sorted(word_importances)
        bottom  = sorted_coeff[:n]
        bottom.reverse()
        top = sorted_coeff[-n:]
        classes[class_idx] = {
            'tops':top,
            'bottom':bottom
        }
    return classes
importance =important_features(cnt_vec,model_cnt,10)
```

```python
def plot_important_words(data, name,n):
    y_pos = np.arange(n)

    top_scores = [a[0] for a in data[1]['tops']]
    top_words = [a[1] for a in data[1]['tops']]
    bottom_scores = [a[0] for a in data[1]['bottom']]
    bottom_words = [a[1] for a in data[1]['bottom']]
    fig = plt.figure(figsize=(10, 10))  

    plt.subplot(121)
    plt.barh(y_pos,bottom_scores, align='center', alpha=0.5,color='red')
    plt.title('Irrelevant', fontsize=20)
    plt.yticks(y_pos, bottom_words, fontsize=14)
    plt.suptitle('Key words', fontsize=16)
    plt.xlabel('Importance', fontsize=20)
    
    plt.subplot(122)
    plt.barh(y_pos,top_scores, align='center', alpha=0.5,color='green')
    plt.title('Disaster', fontsize=20)
    plt.yticks(y_pos, top_words, fontsize=14)
    plt.suptitle(name, fontsize=16)
    plt.xlabel('Importance', fontsize=20)
    
    plt.subplots_adjust(wspace=0.8)
    plt.show()
plot_important_words( importance,"Most important words for relevance",10)
```

Our classifier correctly picks up on some patterns (hiroshima, massacre), but clearly seems to be overfitting on some irellevant terms (sooooo,lordminion777).

# TFIDF Bag of Words

Let's try a slightly more subtle approach. On top of our bag of words model, we use a TF-IDF (Term Frequency, Inverse Document Frequency) which means weighing words by how frequent they are in our dataset, discounting words that are too frequent(the,of etc), as they just add to the noise.

```python
tfidf_vec=TfidfVectorizer()
x_train_tfidf=tfidf_vec.fit_transform(x_train)
x_test_tfidf=tfidf_vec.transform(x_test)
```

```python
plot_LSA(x_train_tfidf,y_train)
```


These embeddings look much more separated, let's see if it leads to better performance.

```python
model_tfidf=LogisticRegression( C=30.0,class_weight='balanced', solver='newton-cg', 
                         multi_class='multinomial', n_jobs=-1, random_state=5)
model_tfidf.fit(x_train_tfidf, y_train)
predicted_tfidf=model_tfidf.predict(x_test_tfidf)
```

```python
print(classification_report(y_test,predicted_tfidf))
print('Accuracy: %.2f%%'%(accuracy_score(y_test,predicted_tfidf)*100))
```

```python
cm = confusion_matrix(y_test,predicted_tfidf)
fig = plt.figure(figsize=(10, 10))
plot = plot_confusion_matrix(cm, classes=['Irrelevant','Disaster','Unsure'],title='Confusion matrix')
plt.show()
```

```python
importance =important_features(tfidf_vec,model_tfidf,10)
plot_important_words( importance,"Most important words for relevance",10)
```

The words it picked up look much more relevant! Although our metrics on our held out validation set haven't increased much,

***This  models have managed to pick up high signal words.However this model don't understand that words like 'good' and 'positive' are closer than apricot and 'continent'.So ,by capturing semantic meaning of a word,this model can improve further .***
